```python
df_train = pd.read_csv('../data/sales_train_validation.csv')
```

```python
columns = ['dept_id', 'cat_id', 'store_id', 'state_id']
[df_train[column].unique() for column in columns]
```

```python
df_train['avg']= df_train[[col for col in df_train if col.startswith('d_')]].mean(axis=1)
```

```python
df_sub = pd.read_csv('../data/sample_submission.csv')
```

```python
df_sub.head()
```

```python
df_sub = df_sub.merge(df_train[['id', 'avg']])
```

```python
df_sub
```

```python
for col in df_sub:
    if col.startswith('F'):
        df_sub[col] = df_sub['avg']
```

```python
df_sub = df_sub.drop('avg', 1)
```

```python
df_sub.to_csv( '../data/submission.csv', index=False)
```
