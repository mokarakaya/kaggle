```python
import pandas as pd
import xgboost as xgb
import numpy as np
```

```python
df = pd.read_csv('../data/train.csv')
```

```python
df.head()
```

```python
X_train = df.drop(columns=['label'])
y_train = df['label']
```

```python
y_train.unique()
```

```python
X_train.columns
```

```python
clf = xgb.XGBClassifier(objective="multi:softmax", num_class=y_train.nunique(), random_state=42)

# Training the classifier
clf.fit(X_train, y_train)

# Making predictions on the testing set
# y_pred = clf.predict(X_test)

# Calculating the accuracy of the classifier
# accuracy = accuracy_score(y_test, y_pred)
```

```python
dft = pd.read_csv('../data/test.csv')
```

```python
predictions = clf.predict(dft)
```

```python
dft["Label"] = predictions
```

```python
dft["ImageId"] = np.arange(1, len(dft) + 1)
```

```python
dft[["ImageId", "Label"]].to_csv("../data/submission.csv", index=False)
```
